```python
import pandas as pd
```

```python
from sklearn.preprocessing import Binarizer 
```

```python
df=pd.read_csv("ingredients.csv")
```

```python
df
```

```python
bin=Binarizer(threshold=5)
```

```python
col=df["sweetness"].values
col=col.reshape(-1,1)
```

```python
col
```

```python
bin.fit_transform(col)
```

```python
df["sweetness"]=bin.fit_transform(col)
```
